### 16. Processing image with different threshold

- 다양한 임계 처리법을 사용해 그레이스케일 이미지를 이진 이미지로 변환하는 방법

1. 모듈을 가져오고 그레이 스케일로 이미지를 읽는다

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('../data/Lena.png', 0)
```

2. 기본 이진 임계 처리를 적용한다

```python
thr, mask = cv2.threshold(img, 200, 1, cv2.THRESH_BINARY)
print('Threshold used: ', thr)
```

3. 적응 임계 처리를 적용한다

```python
adapt_mask = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C,
                                   cv2.THRESH_BINARY_INV, 11, 10)
```

4. 결과를 시각화 한다

```python
plt.figure(figsize=(10,4))
plt.subplot(131)
plt.axis('off')
plt.title('original')
plt.imshow(img, cmap='gray')
plt.subplot(132)
plt.axis('off')
plt.title('binary threshold')
plt.imshow(mask, cmap='gray')
plt.subplot(133)
plt.axis('off')
plt.title('adaptive threshold')
plt.imshow(adapt_mask, cmap='gray')
plt.tight_layout()
plt.show()
```

- OpenCV에는 다양한 유형의 임계값과 임계 처리법이 있다.
- 모든 처리법은 크게 두가지로 나뉠수 있다.
    - 첫번째는 모든 픽셀에 동일한 임계값을 사용하는 전역적 방식이며
    - 두번째는 픽셀에 따라 적용하는 임계값이 다른 적응형 방식이다
- 첫번째 접근 방식은 cv2.threshold 함수를 사용하며, 이 함수는 임계값 타입 (cv2.THRESH_BINARY, cv2.THRESH_BINARY_INV 등이 있음)을 지정하는 매개변수를 받는다
- 적응 임계 처리 방식은 cv2.adaptiveThreshold 함수를 통해 사용한다
- 적응형 방식에서 각 픽셀은 주변 픽셀 값에 따라 독자적인 임계값을 갖는다.
- 앞의 코드에서는 cv2.ADAPTIVE_THRESH_MEAN_C 방식을 사용해 주변 픽셀의 평균에 사용자가 지정한 편향 값(여기서는 10)을 뺀 값을 픽셀 단위의 임계값으로 사용했다.
