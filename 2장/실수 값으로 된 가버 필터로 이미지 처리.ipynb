### 13. Processing image with real-valued Gabor filter

- 가버 필터 커널(이미지의 에지 검출에 용이함)을 구성해 이미지에 적용하는 방법

1. 필요한 모듈을 가져오고 이미지를 그레이 스케일로 읽고 np.float32 데이터 타입으로 변환

```python
import math
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('../data/Lena.png', 0).astype(np.float32) / 255
```

2. 실수 값으로 된 가버 필터 커널을 구성하고, L2 단위 노름을 갖는 방식으로 커널을 정규화함

```python
kernel = cv2.getGaborKernel((21,21), 5, 1, 10, 1, 0 , cv2.CV_32F)
kernel /= math.sqrt((kernel * kernel).sum())
```

3. 이미지를 필터링한다

```python
filtered = cv2.filter2D(img, -1, kernel)
```

4. 결과를 시각화한다

```python
plt.figure(figsize=(8,3))
plt.subplot(131)
plt.axis('off')
plt.title('image')
plt.imshow(img, cmap='gray')
plt.subplot(132)
plt.title('kernel')
plt.imshow(kernel, cmap='gray')
plt.subplot(133)
plt.axis('off')
plt.title('filtered')
plt.imshow(filtered, cmap='gray')
plt.tight_layout()
plt.show()
```

-가버 필터는 코사인파로 변조된 2D 가우시안 커널을 갖는 선형 필터다
- 가버 커널을 얻기 위한 cv2.getGaborKernel 함수는 
- 매개변수로: 커널크기, 가우시안 표준편차, 파형 방향, 파형 길이, 공간비율, 위상을 받는다
- 가버 필터는 이미지의 방향을 아는 경우 에지를 검출하는데 유용한 필터다
