### 11 Computing gradient images using Sobel filter

- 이번 절의 내용은 Sobel 필터를 사용한 이미지 경사도의 계산법이다

1. 필요한 모든 모듈을 가져오고 이미지를 그레이스케일로 읽는다

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('../data/Lena.png', 0)
```

2. Sobel 함수를 사용해 경사도의 근삿값을 계산한다

```python
dx = cv2.Sobel(img, cv2.CV_32F, 1, 0)
dy = cv2.Sobel(img, cv2.CV_32F, 0, 1)
```

3. 결과를 시각화한다

```python
plt.figure(figsize=(8,3))
plt.subplot(131)
plt.axis('off')
plt.title('img')
plt.imshow(img, cmap='gray')
plt.subplot(132)
plt.axis('off')
plt.imshow(dx, cmap='gray')
plt.title(r'$\frac{dI}{dx}$')
plt.subplot(133)
plt.axis('off')
plt.title(r'$\frac{dI}{dy}$')
plt.imshow(dy, cmap='gray')
plt.tight_layout()
plt.show()
```

- OpenCV의 cv2.Sobel 함수는 지정된 크기의 선형 필터를 사용해 이미지 경사도의 근삿값을 계산한다
- 함수의 매개변수를 통해 사용될 커널과 출력 이미지의 데이터 타입, 계산돼야 할 미분 값을 지정할 수 있다
