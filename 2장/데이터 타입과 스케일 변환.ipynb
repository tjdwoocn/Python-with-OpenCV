### 02 Converting between different data types and scaling values

- 행렬 요소의 데이터 타입을 uint8 에서 float32로 변경하고
- 값이 일정 범위 내에 포함되는 것을 고려하지 않고 산술 연산을 수행한 다음에
- 연산결과를 uint8로 변환하는 방법

1. 이미지를 열어 형태와 데이터 타입을 출력하고 이미지를 화면에 표시

```python
import cv2
import numpy as np

image = cv2.imread('../data/Lena.png')
print('Shape: ', image.shape)
print('Data type: ', image.dtype)

cv2.imshow('image', image)
cv2.waitKey()
cv2.destroyAllWindows()
```

2. 이미지를 float 데이터 타입으로 변환

```python
image = image.astype(np.float32)/255
print('Shape: ', image.shape)
print('Data type: ', image.dtype)
```

3. 이미지의 요소를 2배로 스케일링하고 해당 값을 [0,1] 범위로 조절

```python
cv2.imshow('image', np.clip(image*2, 0, 1))
cv2.waitKey()
cv2.destroyAllWindows()
```

4. 이미지의 요소를 다시 [0,255] 범위로 조절하고, 요소 데이터 타입을 uint8 로 변환하기

```python
image = (image*255).astype(np.uint8)
print('Shape: ', image.shape)
print('Data type: ', image.dtype)

cv2.imshow('image', image)
cv2.waitKey()
cv2.destroyAllWindows()
```

- 행렬에 직접 대수 연산을 사용해 행렬 값에 스케일링을 적용할 수 있다
- 예를 들어 행렬을 특정 값(앞의 코드에서는 255)으로 나누면 행렬의 각 요소를 지정된 값으로 나눈다
