### 06 Gamma correlation and per-element math

- 감마 보정은 비선형 방식으로 픽셀 값의 분포를 왜곡하는데 사용
- 감바 보정을 사용하면 이미지를 더 잘 보이게 밝기를 조정할 수 있다

1. 그레이스케일로 이미지를 불러오고 모든 픽셀 값을 [0,1] 범위의 np.float32 데이터 타입으로 변환 

```python
import cv2
import numpy as np

image = cv2.imread('../data/Lena.png', 0).astype(np.float32) / 255
```

2. 지정된 지수 값인 gamma를 사용해 요소당 지수를 적용

```python
gamma = 0.5
corrected_image = np.power(image, gamma)
```

3. 원본과 결과 이미지를 출력한다

```python
cv2.imshow('image', image)
cv2.imshow('corrected_image', corrected_image)
cv2.waitKey()
cv2.destroyAllWindows()
```

- 감마 보정은 이미지 픽셀의 강도를 조정하는 비선형 연산이다
- 이 연산은 입력과 출력 사이의 지수 함수 관계로 표현한다
- 연산은 입력과 출력 사이의 멱법칙 관계 Vout = Vin 으로 표현된다
- 지수 계수가 1보다 큰 값은 이미지를 어둡게 만들고 1보다 작은값은 이미지를 밝게 만든다

```python
gamma = 3.5
corrected_image = np.power(image, gamma)
cv2.imshow('image', image)
cv2.imshow('corrected_image', corrected_image)
cv2.waitKey()
cv2.destroyAllWindows()
```
