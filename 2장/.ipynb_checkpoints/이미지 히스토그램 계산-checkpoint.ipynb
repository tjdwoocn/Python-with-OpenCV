08 Computing image histogram


- 이미지의 히스토그램은 값의 집합에 대한 분포 수준을 보여준다

1. 모듈을 가져오고 이미지를 불러와 화면에 출력

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

grey = cv2.imread('../data/Lena.png', 0)
cv2.imshow('original grey', grey)
cv2.waitKey()
cv2.destroyAllWindows()
```

2. Histogram 함수로 계산한다

```python
hist, bins = np.histogram(grey, 256, [0,255])
```

3. Histogram을 화면에 표시한다

```python
plt.fill(hist)
plt.xlabel('pixel value')
plt.show()
```

- OpenCV에는 히스토그램을 계산하는 내장 함수인 cv.calcHist가 제공됨
- 이번 절에선 코드를 간결하기위해 Numpy를 썻는데 Numpy는 히스토그램을 계산하는
- np.histogram 을 제공한다
- 이 함수는 입력 이미지, 빈의 수, 빈의 범위를 인자로 입력받아 빈의 
- 히스토그램 값과 에지 값을 배열로 반환함
